```python
song_list=[]
for row in data['items']:
    album_id = row['track']['album']['id']
    album_name = row['track']['album']['name']
    album_release_date=row['track']['album']['release_date']
    album_total_tracks=row['track']['album']['total_tracks']
    album_external_url = row['track']['album']['external_urls']['spotify']
    artist_name = row['track']['album']['artists'][0]['name']
    album_element = {'album_id':album_id,'artist':artist_name,'album_name':album_name,'release_date':album_release_date,'total_track':album_total_tracks,'url':album_external_url}
    song_list.append(album_element)
    
    #print(album_element)
    
song_df = pd.DataFrame.from_dict(song_list)
song_df.head()
```
